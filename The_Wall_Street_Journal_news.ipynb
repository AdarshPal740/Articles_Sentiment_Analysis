```python
# In this project we are fetching data of all articles published by the Wall Street Journal in the last 6 months, sorted by recent first with the
# help of newsapi
# The Wall Street  Journal ---->The Wall Street Journal is an American business-focused, international daily newspaper based in New York City,
#  with international editions also available in Chinese and Japanese

# required libraries
import requests
import json
import pandas as pd
```

```python
response=requests.get("https://newsapi.org/v2/everything?domains=wsj.com&apiKey=b01ad54a7594462880c99e46d3cbc242")
response
```

```python
df=pd.DataFrame(response.json()['articles'])# dataframe to fetched data
from google.colab import data_table
data_table.enable_dataframe_formatter() # for data table display
df.head()
```

```python
data=json.loads(response.content) #our data
data['articles'][0] # accessing first articles detail
```

```python
# vaderSentiment is used for Sentiment analysis
! pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# The polarity_scores function will return a dictionary with four values: neg, neu, pos, and compound.
# neg ---> negative value
# pos ---> positive value
# neu ---> neutral value
# compound ---> The Compound score is a metric that calculates the sum of all the lexicon ratings which have been normalized between -1(most extreme negative) and +1 (most extreme positive).
# positive sentiment : (compound score >= 0.05) 
# neutral sentiment : (compound score > -0.05) and (compound score < 0.05) 
# negative sentiment : (compound score <= -0.05)
score = analyzer.polarity_scores(data['articles'][0]['content'])
score
```

**Sentiment Analysis for 4 articles **





```python

for i in range(4):
    score= analyzer.polarity_scores(data['articles'][i]['content'])
    print("\nArticle No :",i+1)
    print("Overall sentiment dictionary is : ", score)
    print("Article was rated as ", score['neg']*100, "% Negative")
    print("Article was rated as ", score['neu']*100, "% Neutral")
    print("Article was rated as ", score['pos']*100, "% Positive")
 
    print("Sentence Overall Rated As", end = " ")
 
    # decide sentiment as positive, negative and neutral
    if score['compound'] >= 0.05 :
        print("Positive")
 
    elif score['compound'] <= - 0.05 :
        print("Negative")
 
    else :
        print("Neutral")
    
```

**Sentiment analysis for whole data**

```python
res=[]

for i in range(len(df)):
    y=()
    y = list(y)
    c=data['articles'][i]['title']
    score= analyzer.polarity_scores(data['articles'][i]['content'])
     # decide sentiment as positive, negative and neutral
    if score['compound'] >= 0.05 :
        y.append('positive')
 
    elif score['compound'] <= - 0.05 :
        y.append('negative')
 
    else :
        y.append('Neutral')
    y=tuple(y)    
    res.append((c,y,score))
Reslut_dataframe= pd.DataFrame(res, columns=['Topic','Overall Sentiment','Sentiment Dictonary'])
```
